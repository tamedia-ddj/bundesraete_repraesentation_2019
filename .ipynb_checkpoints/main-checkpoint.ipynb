```python
# import libraries
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
from datetime import datetime
# import re
# import os
# import json
# import requests
```

```python
# import list of federal council from wikipedia de and wikipedia en
link = "https://de.wikipedia.org/wiki/Liste_der_Mitglieder_des_Schweizerischen_Bundesrates"
bundesraete = pd.read_html(link,header=0)[0]
link = "https://en.wikipedia.org/wiki/List_of_members_of_the_Swiss_Federal_Council"
federal_council = pd.read_html(link, header=0)[0]
```

```python
data = pd.merge(bundesraete, federal_council[["Name", "Election", "Left office", "Party", "Seq[6]"]], on="Name", how="left")
```

```python
data[pd.isna(data.Election)]
```

```python
data.loc[data.Name == "Josef Martin Knüsel", "Election"] = "14 July 1855"
data.loc[data.Name == "Josef Martin Knüsel", "Left office"] = "31 December 1875"
data.loc[data.Name == "Josef Martin Knüsel", "Party"] = "FDP/PRD"
data.loc[data.Name == "Josef Martin Knüsel", "Seq[6]"] = "D02"

data.loc[data.Name == "Marc Ruchet", "Election"] = "14 December 1899"
data.loc[data.Name == "Marc Ruchet", "Left office"] = "9 July 1912"
data.loc[data.Name == "Marc Ruchet", "Party"] = "FDP/PRD"
data.loc[data.Name == "Marc Ruchet", "Seq[6]"] = "G05"
```

```python
data.Election = data.Election.apply(lambda x: datetime.strptime(x, "%d %B %Y"))
data["Left office"] = data["Left office"].apply(lambda x: "amtierend" if (x == "Incumbent") else datetime.strptime(x.split("[")[0], "%d %B %Y"))
```

```python
data.drop(["Bild", "Partei"], axis=1, inplace=True)
```

```python
# Export data to manually edit
data.to_csv("internal/data_edit.csv", index=False)
```
